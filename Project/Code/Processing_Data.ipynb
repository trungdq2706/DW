```python
import pandas as pd
import numpy as np

data = pd.read_csv('../Data/DataCoSupplyChainDataset_Original.csv',encoding=('ISO-8859-1'))
data
```

```python
data.dtypes
```

Xử lý kiểu dữ liệu cột order date (DateOrders) và cột shipping date (DateOrders)

```python
data['order date (DateOrders)'] =  pd.to_datetime(data['order date (DateOrders)'])
data['shipping date (DateOrders)'] =  pd.to_datetime(data['shipping date (DateOrders)'])
```

```python
n = len(data)
for i in range(n):
    order_date = data.loc[i, 'order date (DateOrders)'].date()
    data.loc[i, 'order date (DateOrders)'] = order_date
    
    shipping_date = data.loc[i, 'shipping date (DateOrders)'].date()
    data.loc[i, 'shipping date (DateOrders)'] = shipping_date
```

Tạo file Location.csv

```python
n = len(data)
lst = []
for i in range(n):
    unique_val = data.loc[i, ['Market', 'Order City', 'Order Country', 'Order Region', 'Order State']]
    unique_val_list = unique_val.tolist()
    if len(lst) ==0:
        lst.append(unique_val_list)
    else:
        if unique_val_list not in lst:
            lst.append(unique_val_list)
lst.head()
```

```python
df = pd.DataFrame(lst, columns = ['Market','Order City','Order Country', 'Order Region', 'Order State'])
df['LocationID'] = range(1, len(df) + 1)
```
